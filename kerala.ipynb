in the correlation heatmap ntl mean and ntl sum are highly correlated
If two features (like NTL_Sum and NTL_Mean in your dataset) are highly correlated, they provide similar information to the model, which can lead to issues like multicollinearity—making it difficult to determine the independent effect of each feature and potentially reducing model interpretability and stability

we will apply pca to create a single feature

```python
features_to_combine = ['NTL_Sum', 'NTL_Mean']  # Highly correlated features
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(train_df[features_to_combine])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)  # Combine into 1 component
pca_component = pca.fit_transform(scaled_data)
```

```python
train_df['NTL_PCA'] = pca_component
train_df = train_df.drop(features_to_combine, axis=1)
```

```python
print(f"Variance explained by PCA component: {pca.explained_variance_ratio_[0]:.2%}")
```

 the explained variance ratio for the first principal component (PC1) of your PCA on the NTL_Sum and NTL_Mean features is 95.77%, this means that nearly all the variability in these two features is captured by a single principal component

MODEL TRAINING

RANDOM FOREST

first apply the same pca to test df as well

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select NTL features from test data
test_ntl = test_df[['NTL_Sum', 'NTL_Mean']]

# Standardize and apply PCA (ONLY on test data)
scaler_test = StandardScaler()
ntl_test_scaled = scaler_test.fit_transform(test_ntl)
pca_test = PCA(n_components=1)
test_df['NTL_PCA'] = pca_test.fit_transform(ntl_test_scaled)

# Drop original NTL features
test_df = test_df.drop(['NTL_Sum', 'NTL_Mean'], axis=1)
```

```python
print("Test data columns:", test_df.columns.tolist())
# Output: ['year', 'GDDP', 'District_Number', ..., 'NTL_PCA']
```

```python
# Features (excluding GDDP and district_name)
features = ['year', 'District_Number', 'Population', 'Urbanisation',
            'Urban Population', 'Rural Population', 'pm25_min',
            'pm25_max', 'pm25_mean', 'pm25_num_cells', 'NTL_PCA']

X_train = train_df[features]
y_train = train_df['GDDP']
X_test = test_df[features]
y_test = test_df['GDDP']
```

```python
from sklearn.preprocessing import StandardScaler

# Scale features using training data parameters
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Initialize and fit model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_scaled if 'X_train_scaled' in locals() else X_train, y_train)

# Predict and evaluate on test set
test_pred = rf.predict(X_test_scaled if 'X_test_scaled' in locals() else X_test)
test_rmse = np.sqrt(mean_squared_error(y_test, test_pred))
test_r2 = r2_score(y_test, test_pred)

print(f"Test RMSE: {test_rmse:.2f}")
print(f"Test R²: {test_r2:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# 1. Define parameter grid (expanded from your previous attempt)
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

# 2. Initialize and run Grid Search
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train_scaled, y_train)

# 3. Get best model and parameters
best_rf = grid_search.best_estimator_
best_params = grid_search.best_params_
print(f"Best Parameters: {best_params}")

# 4. Evaluate on test set
test_pred = best_rf.predict(X_test_scaled)
test_rmse = np.sqrt(mean_squared_error(y_test, test_pred))
test_r2 = r2_score(y_test, test_pred)

print(f"\nOptimized Test RMSE: {test_rmse:.2f}")
print(f"Optimized Test R²: {test_r2:.2f}")
```

r2 has improved now we will visualize the results

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Get feature importances from the optimized model
feature_importances = best_rf.feature_importances_

# Create DataFrame for visualization
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
}).sort_values('Importance', ascending=True)

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title('Feature Importance for GDDP Prediction (Optimized Random Forest)')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Initialize and fit XGBoost regressor
xgb = XGBRegressor(n_estimators=200, learning_rate=0.1, max_depth=5, random_state=42)
xgb.fit(X_train_scaled, y_train)

# Predict and evaluate
xgb_pred = xgb.predict(X_test_scaled)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
xgb_r2 = r2_score(y_test, xgb_pred)

print(f"XGBoost Test RMSE: {xgb_rmse:.2f}")
print(f"XGBoost Test R²: {xgb_r2:.2f}")
```
